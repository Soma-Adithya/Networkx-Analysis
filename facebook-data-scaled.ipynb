Feature Extraction

1st feature extraction: Eigen vector centrality.

```python
eigenvector__centrality=nx.eigenvector_centrality(graph)
```

```python
nodes=[]
eigenvector__centralitys=[]
```

```python
for i, j in eigenvector__centrality.items():
    print(i,' : ',j)
    nodes.append(i)
    eigenvector__centralitys.append(j)
```

Z-score normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit_transform(df)
```

```python
df1=pd.DataFrame(scaler)
```

```python
df1.columns=df.columns
```

```python
df1
```

K-means algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.cluster import KMeans
```

```python
MSD = []
for i in range(1,10):
    kmean = KMeans(n_clusters=i)
    kmean.fit(df1)
    MSD.append(kmean.inertia_)
plt.plot(range(1,10),MSD,marker='o')
```

```python
kmean = KMeans(n_clusters=5)
```

```python
kmean.fit(df1)
```

```python
df1["clusters"] = kmean.fit_predict(df1)
```

```python
df1
```

silhouette score

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(df1.drop("clusters",axis=1),df1["clusters"])
```

```python
print(score)
```

calinski harabasz score

```python
from sklearn.metrics import calinski_harabasz_score
calinskiharabaszscore = calinski_harabasz_score(df1.drop("clusters",axis=1),df1["clusters"])
```

```python
calinskiharabaszscore
```

davies bouldin score

```python
from sklearn.metrics import davies_bouldin_score
daviesbouldinscore = davies_bouldin_score(df1.drop("clusters",axis=1),df1["clusters"])
```
